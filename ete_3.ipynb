# PRNP Analysis

```python
from ete3 import Tree

#Input newick file containing phylogeny for aligned sequences
t = ete3.Tree("prion_research/PRNPprotein/aligned_PRNPprotein.tree")
print(t)
```

### Divid susceptible (sep) and resistant species (res) into lists

```python
#Make a list of a susceptible species by leaf
sep_species_PRNP_list = []
for leaf in t.iter_leaf_names():
    if "Sep" in leaf:
        sep_species_PRNP_list.append(leaf)
        
#Make a list of all resistant species by leaf
res_species_PRNP_list = []
for leaf in t.iter_leaf_names():
    if "Res" in leaf:
        res_species_PRNP_list.append(leaf)

#print(f"List of Susceptible Species: {sep_species_PRNP_list}")
#print(f"List of Resistant Species: {res_species_PRNP_list}")
```

### Define a function to get the distance between species

```python
def get_distances(species_list):
    """Return a list of all pairwise distances between species in a species list on a phylogentic tree. 
    Example: pairwise_distances = get_distances(available_list)"""

    distances = []
    
    species_pairs = [(a, b) for idx, a in enumerate(species_list) for b in species_list[idx + 1:]]
    for a, b in species_pairs:
        pairwise_distance = t.get_distance(a,b)
        distances.append(pairwise_distance)
    
    return distances
```

### Apply the distance function

```python
from statistics import mean

#Call on the sep species list and apply the distance function
sep_distances = get_distances(sep_species_PRNP_list)
#Average the distance
average_sep_distance = mean(sep_distances)
print(f"Average distance between susceptible species:", average_sep_distance)

#Call on the res species list and apply the distance function
res_distances = get_distances(res_species_PRNP_list)
#Average the distance
average_res_distance = mean(res_distances)
print(f"Average distance between resistant species:", average_res_distance)
```
